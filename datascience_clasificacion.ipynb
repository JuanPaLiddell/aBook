```
import pandas as pd

df = pd.read_excel('quejas medicas.xlsx')

df.head()
```

```
# Eliminar las columnas 'Fecha_recepcion' y 'Fecha_creacion'
#df = df.drop(['Fecha_recepcion', 'Fecha_creacion', '	Fecha_vencimiento' ], axis=1)

# Verificar el DataFrame actualizado
df.head()
```

```
distinctive_words = [' Norte', 'UMAE Esp. CMN La Raza, ', 'UMAE Esp. CMN Siglo XXI, ', 'UMAE Trauma y Ortopedia Magdalena de las Salinas, ', 'UMAE Esp. CMNO, ', 'UMAE Trauma y Ortopedia CMN, ']

for word in distinctive_words:
    df['OOAD'] = df['OOAD'].str.replace(word, '')

    df.head()
```

```
df
```

```
import matplotlib.pyplot as plt

# Contar el número de quejas por estado (valor en el campo 'OOAD')
quejas_por_estado = df['OOAD'].value_counts()

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
quejas_por_estado.plot(kind='bar', color='skyblue')
plt.title('Número de Quejas Médicas por Estado (OOAD)')
plt.xlabel('Estado')
plt.ylabel('Número de Quejas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```

# Contar el número de veces que se repite cada valor en el campo 'Especialidad'
especialidades_count = df['Especialidad'].value_counts()

# Crear la gráfica de barras horizontal
plt.figure(figsize=(10, 6))
especialidades_count.plot(kind='barh', color='lightgreen')
plt.title('Frecuencia de Especialidades en Quejas Médicas')
plt.xlabel('Número de Quejas')
plt.ylabel('Especialidad')
plt.tight_layout()
plt.show()
```

```
# Contar el número de veces que se repite cada valor en el campo 'Daño_agraviado'
dano_agraviado_count = df['Dano_agraviado'].value_counts()

# Crear la gráfica de barras horizontal
plt.figure(figsize=(10, 6))
dano_agraviado_count.plot(kind='barh', color='lightcoral')
plt.title('Frecuencia de Daño Agraviado en Quejas Médicas')
plt.xlabel('Número de Quejas')
plt.ylabel('Daño Agraviado')
plt.tight_layout()
plt.show()
```
